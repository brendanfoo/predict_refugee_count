```python
data = pd.read_csv ('/content/drive/MyDrive/210RefugeeMoves/colab/idmc/IDMC_GIDD_Disasters_Internal_Displacement_Data.csv')
data = data.drop(['Hazard Category', 'Hazard Sub Type', 'Disaster Internal Displacements (Raw)'], axis=1)
```

```python
data.head(5)
```

```python
data = data.loc[data['Year'] >= 2016]
data = data.loc[data['Hazard Type'] == 'Flood']
data = data.loc[data['Country / Territory'] != 'China']
data = data.loc[data['Country / Territory'] != 'India']
data.sort_values('Disaster Internal Displacements', ascending=False).head(20)
```

```python
data.groupby(by=['Hazard Type'])['Disaster Internal Displacements'].count().sort_values(ascending=False)
```

```python
data.groupby(by=['Hazard Type'])['Disaster Internal Displacements'].sum().sort_values(ascending=False)
```

```python
grouped = data.groupby(by=['Hazard Type']).agg({'Disaster Internal Displacements': ['count', 'sum', 'mean']})
grouped.sort_values([('Disaster Internal Displacements', 'mean')], ascending=False)
```
